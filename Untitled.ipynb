```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("stats.csv", newline="") as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     temp = list(reader)
     drivers = [int(x[1]) for x in temp[1:]]
     min_speed = [float(x[2]) for x in temp[1:]]
     avg_speed = [float(x[3]) for x in temp[1:]]
     max_speed = [float(x[4]) for x in temp[1:]]
```

```python
avg_speed
```

```python
plt.plot(drivers)
plt.yticks([0, 20, 40, 60, 80, 100], [0, 20, 40, 60, 80, 100])
plt.rcParams["figure.figsize"] = (10,2)
plt.ylabel("Car count")
plt.show()
```

```python
plt.plot(min_speed)
plt.ylabel("Min speed")
plt.rcParams["figure.figsize"] = (10,2)
plt.yticks(np.arange(0, 10, step=1))
plt.show()
```

```python
plt.plot(avg_speed)
plt.ylabel("Avg speed")
plt.yticks([0, 2, 4, 6, 8, 10], [0, 2, 4, 6, 8, 10])
plt.show()
```

```python
plt.plot(max_speed)
plt.ylabel("Max speed")
plt.yticks([0, 2, 4, 6, 8, 10], [0, 2, 4, 6, 8, 10])
plt.show()
```
